برای انجام مرحله Unroll کردن آرایه تصویر به صورت حرفه‌ای و بهینه، باید مطمئن شویم که آرایه‌ی تصویر به درستی به یک آرایه دوبعدی تبدیل شود، به طوری که هر ردیف از این آرایه جدید، مقادیر کانال‌های رنگی (RGB) مربوط به یک پیکسل خاص را در خود داشته باشد.

مرحله 1: Unroll کردن آرایه
برای این کار، به صورت زیر عمل می‌کنیم:

تبدیل آرایه سه‌بعدی تصویر به یک آرایه دوبعدی: از تابع reshape در NumPy استفاده می‌کنیم تا شکل آرایه‌ی تصویر را تغییر دهیم. تعداد پیکسل‌ها در تصویر برابر با حاصل‌ضرب طول و عرض تصویر است. هر پیکسل نیز یک بردار با سه مقدار (RGB) خواهد بود.
نمایش نتایج: پس از Unroll کردن آرایه، شکل جدید آرایه و چند ردیف از آن را برای اطمینان از صحت عملیات چاپ می‌کنیم.

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt


image_path = 'pepp.tif'
image = Image.open(image_path)
image_array = np.array(image)


# Reshape and unroll
pixels = image_array.reshape((-1, 3))

print("unroll", pixels.shape)
print(pixels[:100])

plt.imshow(image_array)
plt.axis('off')
plt.show()
```

توضیحات و جزئیات کد:
باز کردن تصویر: تصویر با استفاده از PIL.Image.open باز شده و به یک آرایه نام‌پای تبدیل می‌شود.
Unroll کردن آرایه: با استفاده از reshape، آرایه‌ی سه‌بعدی تصویر (که دارای ابعاد (ارتفاع، عرض، ۳) است) به یک آرایه دوبعدی با ابعاد (تعداد پیکسل‌ها، ۳) تبدیل می‌شود. در اینجا، تعداد پیکسل‌ها برابر است با ارتفاع * عرض.
چاپ شکل و نمونه‌های آرایه: پس از Unroll کردن آرایه، شکل آن ((تعداد پیکسل‌ها، ۳)) چاپ شده و چند نمونه از ردیف‌های اول آن به نمایش درمی‌آید تا مطمئن شویم هر ردیف شامل یک پیکسل است که دارای مقادیر RGB می‌باشد.
نمایش تصویر: تصویر اولیه برای اطمینان از اینکه تصویر اصلی هنوز به درستی نمایش داده می‌شود، نمایش داده می‌شود.
